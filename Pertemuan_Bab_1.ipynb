Rangkuman Bab 1: Review Konversi Entity Relationship (ER) Diagram ke Skema Relasi
Bab 1 membahas proses konversi model konseptual data yaitu Entity Relationship Diagram (ERD) menjadi skema relasi yang dapat diimplementasikan ke dalam basis data relasional. Ini adalah langkah awal yang penting sebelum membuat database fisik di komputer.

Pokok-pokok Materi:
Konversi Entitas Kuat
Setiap entitas kuat menjadi sebuah tabel, dengan setiap atribut menjadi kolom, dan atribut kunci menjadi primary key tabel tersebut.

Konversi Atribut Komposit
Atribut komposit yang terdiri dari beberapa bagian sederhana bisa dipecah menjadi beberapa kolom di tabel entitas.

Konversi Atribut Multivalue
Atribut multivalue diubah menjadi tabel terpisah, yang biasanya berisi primary key entitas dan nilai atribut multivalue sebagai kolom.

Konversi Atribut Derivate
Atribut yang dihitung (derivate) tetap menjadi kolom dalam tabel.

Konversi Entitas Lemah
Entitas lemah menjadi tabel sendiri dengan primary key yang mengandung primary key entitas kuat sebagai foreign key.

Konversi Relasi

Relasi satu-ke-satu (1:1) biasanya diterapkan dengan menambahkan foreign key di salah satu tabel entitas.

Relasi satu-ke-banyak (1:N) dibuat dengan menempatkan foreign key di tabel entitas yang merupakan sisi "banyak".

Relasi banyak-ke-banyak (M:N) harus dipisah menjadi tabel relasi baru yang memuat foreign key dari kedua entitas.

Relasi dengan Atribut
Jika relasi memiliki atribut, atribut tersebut menjadi kolom di tabel relasi baru yang dibentuk.

Relasi Ternary dan Unary
Relasi dengan banyak entitas atau self-join (relasi dengan entitas yang sama) dijelaskan dengan contoh skema tabel terkait.

Generalisasi dan Spesialisasi
Pendekatan pembuatan tabel untuk subclass dan superclass entitas dengan metode yang berbeda.

Agregasi
Penjelasan bagaimana entitas agregasi direpresentasikan dalam basis data relasional.

Studi Kasus
Bab ini juga menyertakan studi kasus konversi ERD skema apotik menjadi tabel-tabel yang saling berelasi lengkap dengan primary key dan foreign key.

Tujuan
