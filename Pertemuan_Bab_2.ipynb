# Rangkuman Bab 2: Pengantar Basis Data - DDL

Bab 2 mengupas pengenalan dasar tentang basis data dan penggunaan awal MySQL sebagai DBMS untuk manajemen basis data relasional.

## Pokok-pokok Materi:

- **Pengertian DBMS dan MySQL**  
  DBMS (Database Management System) adalah sistem pengelolaan basis data menggunakan bahasa query, seperti SQL. MySQL adalah salah satu DBMS populer yang open source, cepat, dapat dipercaya, dan mendukung berbagai platform (Linux, Windows, Mac OS).

- **Instalasi MySQL**  
  Instalasi dapat dilakukan bersamaan dengan paket XAMPP yang juga menyertakan phpMyAdmin. Versi terkini XAMPP menggunakan MariaDB yang kompatibel dengan MySQL.

- **Server MySQL dan Client**  
  MySQL Server adalah engine yang menyimpan database, berjalan di latar belakang. Client MySQL (mysql.exe) digunakan untuk mengakses dan mengelola basis data melalui command line.

- **Mengakses MySQL**  
  Pengguna dapat login MySQL melalui terminal menggunakan perintah seperti `mysql -u root -p` dan mengelola database dengan query SQL.

- **Tipe Data MySQL**  
  Beragam tipe data disediakan antara lain INT (angka bulat), FLOAT (angka desimal), DATE, DATETIME, CHAR, VARCHAR (string), BLOB, dan LONG BLOB untuk menyimpan berbagai bentuk data.

- **Basis Data Relasional**  
  Database terdiri dari kumpulan tabel yang terdiri dari baris dan kolom. Definisi database harus dibuat dan diaktifkan sebelum membuat tabel di dalamnya.

- **Data Definition Language (DDL)**  
  Perintah DDL digunakan untuk mendefinisikan dan mengelola struktur basis data, khususnya database dan tabel. Contoh perintah DDL:  
  - `CREATE DATABASE nama_database;` untuk membuat database baru  
  - `SHOW DATABASES;` untuk melihat daftar database  
  - `USE nama_database;` untuk memilih database aktif  
  - `DROP DATABASE nama_database;` untuk menghapus database

## Tujuan Kompetensi  
Praktikan dapat memahami konsep dasar basis data, menginstall dan menggunakan MySQL, mengakses dan membuat database menggunakan perintah DDL.
